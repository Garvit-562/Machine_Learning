```python
import pandas as pd
import numpy as np

```

```python
df = pd.read_csv('Unicorn_Companies.csv')
```

```python

df['Valuation ($B)'] = (
    df['Valuation']
    .str.replace('$', '', regex=False)
    .str.replace('B', '', regex=False)
    .astype(float)
)
df['Funding_clean'] = (
    df['Funding']
    .str.replace('$', '', regex=False)
    .str.replace('B', '', regex=False)
    .str.replace('M', '', regex=False)
    .replace('Unknown', pd.NA)
)

df['Funding ($B)'] = pd.to_numeric(df['Funding_clean'], errors='coerce')

is_million = df['Funding'].str.contains('M', na=False)
df.loc[is_million, 'Funding ($B)'] = df.loc[is_million, 'Funding ($B)'] / 1000

# Drop helper column
df.drop(columns='Funding_clean', inplace=True)
df.head(10)
```

```python
#calculating funding/valuation ratio by converting Funding($B) and Valuation($B) columns to arrays

valuation = df['Valuation ($B)'].to_numpy()
funding = df['Funding ($B)'].to_numpy()

ratio = np.where(valuation > 0, funding / valuation, np.nan)
df['Funding-to-Valuation Ratio'] = ratio
df.head(10)
```

```python
#adding column to see how long company took to become a unicorn

df['Date Joined'] = pd.to_datetime(df['Date Joined'], errors='coerce')
df['Join Year'] = df['Date Joined'].dt.year
year_founded = df['Year Founded'].to_numpy()
join_year = df['Join Year'].to_numpy()

df['Years to Unicorn'] = join_year - year_founded
df.head(10)
```

```python
#categorizing each company based on bins such that ; valuation<5 = 0 , 5<valuation<10 = 1, valuation>10 = 2

valuation_bins = np.digitize(valuation, bins=[5, 10])
df['Valuation Bucket'] = valuation_bins
df.groupby('Valuation Bucket').size()
```

```python

valuation_mean = np.nanmean(valuation)
valuation_std = np.nanstd(valuation)
funding_mean = np.nanmean(funding)
funding_std = np.nanstd(funding)

df['Valuation Z'] = (valuation - valuation_mean) / valuation_std
df['Funding Z'] = (funding - funding_mean) / funding_std
print(df[['Valuation Z','Funding Z']])
```

```python
valuation = df['Valuation ($B)'].to_numpy()
funding = df['Funding ($B)'].to_numpy()
years = df['Years to Unicorn'].to_numpy()

# Calculate CAGR using NumPy (safe handling of division, invalid years)
with np.errstate(divide='ignore', invalid='ignore'):
    cagr = np.where(
        (funding > 0) & (valuation > 0) & (years > 0),
        (valuation / funding) ** (1 / years) - 1,
        np.nan
    )

# Store the result
df['CAGR'] = cagr

# Display a sample
df[['Company', 'Valuation ($B)', 'Funding ($B)', 'Years to Unicorn', 'CAGR']].head(10)
```

```python
#Calculating Efficiency score that tells you how much valuation each company gained per dollar per year  a higher score means the company became a unicorn more efficiently
valuation = df['Valuation ($B)'].to_numpy()
funding = df['Funding ($B)'].to_numpy()
years = df['Years to Unicorn'].to_numpy()

# Compute efficiency score with NumPy
with np.errstate(divide='ignore', invalid='ignore'):
    efficiency_score = np.where(
        (funding > 0) & (years > 0),
        valuation / (funding * years),
        np.nan
    )

# Add to DataFrame
df['Efficiency Score'] = efficiency_score

# Show top 10 most efficient unicorns
df[['Company', 'Valuation ($B)', 'Funding ($B)', 'Years to Unicorn', 'Efficiency Score']]\
  .sort_values(by='Efficiency Score', ascending=False).head(10)
```
